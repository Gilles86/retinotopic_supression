```python
import pandas as pd
import numpy as np
```

```python
n_trials = 5

possible_orientations = [0, np.pi/2]
possible_colors = ['red', 'green']
possible_locations = range(0, 8)


df = []
for trial in range(n_trials):
    target_orientation = np.random.choice(possible_orientations)
    singleton_color = np.random.choice(possible_colors)
    target_location = np.random.choice(possible_locations)
    # Cannot be target location
    singleton_location = np.random.choice([loc for loc in possible_locations if loc != target_location])

    df.append({
        'trial': trial,
        'target_orientation': target_orientation,
        'singleton_color': singleton_color,
        'target_location': target_location,
        'singleton_location': singleton_location
    })

df = pd.DataFrame(df)
display(df)

df.to_csv('settings/test.tsv', sep='\t', index=False)
```
