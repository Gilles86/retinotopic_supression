```python
from nilearn import image
import nibabel as nib
from pymp2rage import MP2RAGE
```

```python
img1 = image.load_img('/Users/gdehol/data/ds-retsupp/sourcedata/mri/sub-01/ses-2/RetSup_testS2s2_acq-MP2_085mm_6_1.PAR')

img1_nii = nib.Nifti1Image(img1.get_fdata(), img1.affine, img1.header)
```

```python
inv1 = image.index_img(img1_nii, 0)
inv2 = image.index_img(img1_nii, 1)
inv1_phase = image.index_img(img1_nii, 2)
inv2_phase = image.index_img(img1_nii, 3)
```

```python
MP2RAGE_obj = MP2RAGE(MPRAGE_tr=4.2, invtimesAB=[0.7, 1.5], inv1=inv1, inv2=inv2, inv1ph=inv1_phase, inv2ph=inv2_phase)
```

```python
from utils import create_preprocessed_t1w
```

```python
t1w = MP2RAGE_obj.fit_t1w_uni()

image.math_img('np.nan_to_num(t1w)', t1w=t1w)

t1w.to_filename('/Users/gdehol/data/ds-retsupp/sourcedata/mri/sub-01/ses-2/t1w.nii.gz')
inv2.to_filename('/Users/gdehol/data/ds-retsupp/sourcedata/mri/sub-01/ses-2/inv2.nii.gz')
```
