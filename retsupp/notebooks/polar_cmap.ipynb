```python
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(-np.pi, np.pi, 1000)
r = np.linspace(0, 1, 100)
Theta, R = np.meshgrid(theta, r)
Z = Theta

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
fig.patch.set_facecolor('black')
ax.patch.set_facecolor('black')

# Plot the data
c = ax.pcolormesh(Theta, R, Z, cmap='hsv', vmin=-np.pi, vmax=np.pi)

# Adjust radial limits for label space
ax.set_rlim(0, 1.025)

# Set xticks to white
ax.tick_params(colors='white', labelsize=18, pad=10)

# Rotate xtick labels to be orthogonal (radial)
xtick_labels = ax.get_xticklabels()
xtick_angles = ax.get_xticks()

plt.yticks([])  # Remove yticks

plt.savefig(
    target_dir / 'polar_cmap.pdf',
    dpi=300,
    bbox_inches='tight',
    transparent=False,
    facecolor=fig.get_facecolor()
)
```
